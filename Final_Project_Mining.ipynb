# **Data preprocessing**

```python
def preprocessing(text):
    text = text.lower()
    text = re.sub("[^a-z \:]+", "", text)
    result = []
    for word in text.split():
        if word not in stop_words:
            word = wnl.lemmatize(word, "n")
            word = wnl.lemmatize(word, "v")
            word = wnl.lemmatize(word, "a")
            word = stemmer.stem(word)
            result.append(word)
    return " ".join(result)
```

```python
import pandas as pd
import re
import json

# General Preprocessing Function (lowercasing, remove punctuation)
def preprocess(text):
    return re.sub(r'\W', ' ', text.lower())

# Relevance Function - Checks if the query matches the document based on keyword overlap
def is_relevant(query, document, overlap_threshold=2):
    query_keywords = set(preprocess(query).split())
    document_text = preprocess(document)
    
    # Tokenize the document into words and check overlap
    document_keywords = set(document_text.split())
    
    # Calculate overlap between the query and the document
    overlap = query_keywords.intersection(document_keywords)
    
    # Consider the document relevant if the overlap exceeds a threshold
    if len(overlap) >= overlap_threshold:  
        return True
    return False

# General function to generate ground truth
def generate_ground_truth(queries, dataset, text_fields=['description'], overlap_threshold=2):
    """
    Generate ground truth relevance for a set of queries against a dataset.
    
    Args:
        queries (list): List of queries to search for relevance.
        dataset (pd.DataFrame): The dataset containing documents (each row is a document).
        text_fields (list): List of columns to use for document text (e.g., 'description', 'title').
        overlap_threshold (int): Number of keyword overlaps to consider the document relevant.
    
    Returns:
        dict: A dictionary where the key is the query and the value is a list of relevant document IDs.
    """
    ground_truth = {}

    # Iterate through each query
    for query in queries:
        relevant_docs = []
        
        # Iterate through each document in the dataset
        for _, row in dataset.iterrows():
            doc_id = row['docno']
            
            # Check relevance in any of the given fields
            for field in text_fields:
                if field in row and is_relevant(query, row[field], overlap_threshold):
                    relevant_docs.append(doc_id)
                    break  # Stop once we find one relevant field for the document
        
        # Store the relevant documents for the query
        ground_truth[query] = relevant_docs
    
    return ground_truth

# Example Usage
# Assume `df2` is the dataset and has columns like 'docno', 'description', 'title', etc.

# List of queries (modify as per your need)
queries = ['laptop', 'smartphone', 'headphones', 'tablet', 'gaming']

# Generate ground truth
ground_truth = generate_ground_truth(queries, df_first, text_fields=['object', 'description'], overlap_threshold=1)

# Print ground truth
for query, relevant_docs in ground_truth.items():
    print(f"Query: '{query}'")
    print(f"Relevant Documents: {relevant_docs}")
    print()

# Optionally, save the ground truth to a file
with open('ground_truth.json', 'w') as f:
    json.dump(ground_truth, f)
```

```python
def calculate_metrics(retrieved_docs, query):
    # Get the relevant documents for the current query
    relevant_docs = generate_ground_truth(query).get(query, [])
    
    if not relevant_docs:
        return {'precision': 0, 'recall': 0, 'f1_score': 0}  # No relevant documents available
    
    # Calculate Precision, Recall, and F1-Score
    retrieved_set = set(retrieved_docs)
    relevant_set = set(relevant_docs)
    
    true_positives = len(retrieved_set & relevant_set)
    false_positives = len(retrieved_set - relevant_set)
    false_negatives = len(relevant_set - retrieved_set)
    
    precision = true_positives / (true_positives + false_positives) if true_positives + false_positives > 0 else 0
    recall = true_positives / (true_positives + false_negatives) if true_positives + false_negatives > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if precision + recall > 0 else 0
    
    return {'precision': precision, 'recall': recall, 'f1_score': f1}
```

```python
import os
df["docno"] = list(range(df.shape[0]))
df["docno"] = df["docno"].astype(str)
docs = df.to_dict("records")

index_path = os.path.abspath("myIndex2")
os.makedirs(index_path, exist_ok=True)

indexer = pt.IterDictIndexer(index_path, fields=["text"], meta=["docno"])
index_ref = indexer.index(docs)
index = pt.IndexFactory.of(index_ref)

```

```python
def search_term(term_original):

    term = preprocessing(term_original)
    print(f"\nSearching for: {term_original}")
    print(f"Term after preprocessing is: {term}")


    try:
        pointer = index.getLexicon()[term]
        postings_id=[]
        for posting in index.getInvertedIndex().getPostings(pointer):
           postings_id.append(str(posting.getId()))
        print(f"Term {term} was found in {len(postings_id)} documents with document IDs {postings_id}\nThese documents are:")
        for posting in index.getInvertedIndex().getPostings(pointer):
          id = posting.getId()
          details = df2[df2['docno'] == str(id)]["text"].values[0]
          print(f"\nDocument ID: {id}, Length: {posting.getDocumentLength()}\nDetails: {details}")

    except KeyError:
        print(f" Term '{term}' not found in the index.")
```

# **Query expansion**

```python
def expand_query_rm3(query, model="BM25"):
    retr = pt.BatchRetrieve(index_ref, wmodel=model)

    pipeline = retr >> pt.rewrite.RM3(index_ref)

    query_df = pd.DataFrame([{"qid": "1", "query": query}])

    result_df = pipeline.transform(query_df)

    expanded_query = result_df.iloc[0]["query"]
    return expanded_query

# Example usage:
print(expand_query_rm3("car"))
```
